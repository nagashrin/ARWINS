```python
import pandas as pd
import numpy as np
import time, os, fnmatch, shutil

#Create variable timestamp 
t = time.localtime()
timestamp = time.strftime('%Y%m%d_%H%M', t)
```

```python
# load file
aries_df = pd.read_excel('S://TANF Reports//!2022 TANF Reports//test//data//ARIES_PCRLISTFORDWS20220405.xlsx')

# fill na with 0
aries_df = aries_df.fillna(0)

# convert Date_OF_BIRTH column to date time format
aries_df['DATE_OF_BIRTH'] = pd.to_datetime(aries_df['DATE_OF_BIRTH'], infer_datetime_format=True)

# convert RES_ADDRESS_CITY column to title format
aries_df['RES_ADDRESS_CITY'] = aries_df['RES_ADDRESS_CITY'].str.title()

# create a copy of file
aries = aries_df.copy()
```

```python
aries = aries_df.copy()
```

```python
# create a subset of clients having email address
aries_email = aries.loc[aries['EMAIL_ADDRESS'] != 0]


# groupby clients having email address and zip and count of ssn
aries_email_zip = aries_email.groupby('RES_ADDRESS_ZIPCODE').agg({'SSN':'count'}, axis = 'columns').rename({'SSN': 'count'}).reset_index()
aries_email_zip = aries_email_zip.sort_values(by=['SSN'], ascending = False) 

# groupby clients having email address and city and count of ssn
aries_email_city = aries_email.groupby('RES_ADDRESS_CITY').agg({'SSN':'count'}, axis = 'columns').rename({'SSN': 'count'}).reset_index()
aries_email_city = aries_email_city.sort_values(by=['SSN'], ascending = False) 
```

```python
aries_email_phone = aries[aries['TELEPHONE_NUMBER']!= 0].shape
aries_email_phone
```

```python
os.chdir('S://TANF Reports//!2022 TANF Reports//test')
os.getcwd()


file_name = 'ARIES-DWS_' + timestamp +'.xlsx'
writer = pd.ExcelWriter(file_name, engine='xlsxwriter')

aries_email_zip.to_excel(writer, sheet_name = 'email-zip', index=False)
aries_email_city.to_excel(writer, sheet_name = 'email_city', index = False)

writer.save()
```
